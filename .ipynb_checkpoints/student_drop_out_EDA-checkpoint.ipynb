```python
# Importing  libraries and visualisation tools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
'''
plt.figure(figsize=(10,10))
plt.title('Academic Status')
plt.pie['Target'].value_counts(),labels = ['Graduate','Dropout','Emrolled']}
plt.legend(loc = 'lower right')
#sns.barplot(x='Target', y='Gender', data=df)
'''
target_counts = df['Target'].value_counts()

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title("Distribution of Target Variable (Graduate vs. Dropout)")
plt.show()
```

```python
# Calculate the value counts for each class in the target variable
target_counts = df['Target'].value_counts()
labels = target_counts.index  # Labels for the pie chart (e.g., 'Graduate', 'Dropout')
sizes = target_counts.values   # Values for each class

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title('Distribution of Target Variable (Graduate vs Dropout)')
plt.show()
```

```python

# Plot a bar chart for gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette='pastel')

plt.xticks([1, 0], ['Male', 'Female'])

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Plot a bar chart for gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming 'Gender' and 'Target' columns exist in your DataFrame
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Target', data=df, palette='Set2')
plt.xticks([1, 0], ['Male', 'Female'])
plt.title('Gender Distribution by Target')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Target')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Example: Encode the 'Target' column
label_encoder = LabelEncoder()
df['Target'] = label_encoder.fit_transform(df['Target'])

# Now, you can check the DataFrame to see that the 'Target' column is numeric
print(df['Target'].head())
```

```python
# Example: One-Hot Encoding for categorical columns
df_encoded = pd.get_dummies(df, columns=['Marital status', 'Application mode', 'Course', 'Previous qualification'], drop_first=True)

# Now check the encoded DataFrame
print(df_encoded.head())
```

```python
correlation_matrix = df.corr(method='pearson')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Pearson Correlation Matrix')
plt.show()
```

```python
# using pearson correlation to compare between features 
correlation_value = df['GDP'].corr(df['Unemployment rate'], method='pearson')
print(f"Pearson correlation between GDP and Unemployment rate: {correlation_value}")
```

```python

# Assuming 'df' is your DataFrame
# Calculate correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_with_target = df.corr(method='pearson')['Target']

# Display the correlation values with the target feature
print(correlation_with_target)
```

```python
# using spearman correlation ( non linear)
# Calculate Spearman correlation between all features and the target
correlation_with_target_spearman = df.corr(method='spearman')['Target']

# Display the correlation values with the target feature
print(correlation_with_target_spearman)
```

```python
# Scatter plot of two variables, say 'Feature1' and 'Target'
plt.scatter(df['Curricular units 2nd sem (approved)'], df['Target'])
plt.title('Scatter Plot: Feature1 vs Target')
plt.xlabel('Feature1')
plt.ylabel('Target')
plt.show()
```
